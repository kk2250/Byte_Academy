```python
df = pd.read_csv('football_2.csv')
```

```python
#df = df.drop(['Rk'], axis=1)
df.head()
```

```python
def name_split(name):
    name = name.split('\\')[0]
    return name
df['Unnamed: 1'] = df.apply(lambda row: name_split(row['Unnamed: 1']), axis=1)
```

```python
def name_strip(name):
    name = name.strip('*+\\')
    return name
df['Unnamed: 1'] = df.apply(lambda row: name_strip(row['Unnamed: 1']), axis=1)
```

```python
df.head(3)
```

```python
df['pts'] = (df['TD']*4)+(df['Yds']*.04)-(df['Yds.1']*.1)
```

```python
df['ppg'] = (df['pts']/df['G']).round()
```

```python
df = df.sort_values(by=['ppg','Cmp%'], ascending=False)
```

```python
df_top_n = df.loc[(df['ppg']>16)]
```
