```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
# using matplotlib in jupyter notebook
```

```python
df = pd.DataFrame(np.random.randn(500, 4), index=pd.date_range('1/1/2000', periods=500), columns=list('ABCD'))
```

```python
df.head()
```

```python
df = df.cumsum()
# cumulative sum - adding index[n]+index[n+1]
```

```python
df.plot()
df.plot(legend=False)
```
